```python
# Dependencies
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd
import os
```

```python
# Import CSV
data = pd.read_csv('Car_safety.csv')
```

```python
car_data = pd.DataFrame(data)
```

```python
# Data contain non-numerical values need to convert 
car_data.head(10)
```

```python
list(car_data.columns.values)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
number = LabelEncoder()
car_data['Buying'] = number.fit_transform(car_data['Buying'].astype('str'))
car_data['Maintenance'] = number.fit_transform(car_data['Maintenance'].astype('str'))
car_data[' Doors'] = number.fit_transform(car_data[' Doors'].astype('str'))
car_data[' Persons'] = number.fit_transform(car_data[' Persons'].astype('str'))
car_data[' Lug_Boot'] = number.fit_transform(car_data[' Lug_Boot'].astype('str'))
car_data[' Safety'] = number.fit_transform(car_data[' Safety'].astype('str'))
car_data[' Class_Values'] = number.fit_transform(car_data[' Class_Values'].astype('str'))
```

```python
car_data.to_csv('car_data_numeric.csv')
```
